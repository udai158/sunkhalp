### AQI calculation for a given dataset

```python
# Function to calculate SO2 individual pollutent index (SO2i):
def calculate_SO2i(SO2):
    SO2i = 0
    if (SO2<=40):
        SO2i = SO2*(50/40)
    if(SO2>40 and SO2<=80):
        SO2i = 50+(SO2-40)*(50/40)
    if(SO2>80 and SO2<=380):
        SO2i = 100+(SO2-80)*(100/300)
    if(SO2>380 and SO2<=800):
        SO2i = 200+(SO2-380)*(100/(800-380))
    if(SO2>800 and SO2<=1600):
        SO2i = 300+(SO2-800)*(100/800)
    if(SO2>1600):
        SO2i = 400 + (SO2-1600)*(100/800)
    return SO2i
```

```python
# Function to calculate PM10 individual pollutent index (PM10i):
def calculate_PM10i(PM10):
    PM10i = 0
    if (PM10<=50):
        PM10i = PM10*(50/50)
    if(PM10>50 and PM10<=100):
        PM10i = 50+(PM10-50)*(50/50)
    if(PM10>100 and PM10<=250):
        PM10i = 100+(PM10-100)*(100/150)
    if(PM10>250 and PM10<350):
        PM10i = 200+(PM10-250)*(100/(100))
    if(PM10>350 and PM10<=430):
        PM10i = 300+(PM10-350)*(100/80)
    if(PM10>430):
        PM10i = 400 + (PM10-430)*(100/80)
    return PM10i
```

```python
# Function to calculate PM2_5 individual pollutent index (PM2_5i):
def calculate_PM2_5i(PM2_5):
    PM2_5i = 0
    if (PM2_5<=30):
        PM2_5i = PM2_5*(50/30)
    if(PM2_5>30 and PM2_5<=60):
        PM2_5i = 50+(PM2_5-30)*(50/30)
    if(PM2_5>60 and PM2_5<=90):
        PM2_5i = 100+(PM2_5-60)*(100/30)
    if(PM2_5>90 and PM2_5<120):
        PM2_5i = 200+(PM2_5-90)*(100/(30))
    if(PM2_5>120 and PM2_5<=250):
        PM2_5i = 300+(PM2_5-120)*(100/130)
    if(PM2_5>250):
        PM2_5i = 400 + (PM2_5-250)*(100/130)
    return PM2_5i
```

```python
# Function to calculate NO2 individual pollutent index (NO2i):
def calculate_NO2i(NO2):
    NO2i = 0
    if (NO2<=40):
        NO2i = NO2*(50/40)
    if(NO2>40 and NO2<=80):
        NO2i = 50+(NO2-40)*(50/40)
    if(NO2>80 and NO2<=180):
        NO2i = 100+(NO2-80)*(100/100)
    if(NO2>180 and NO2<280):
        NO2i = 200+(NO2-180)*(100/(100))
    if(NO2>280 and NO2<=400):
        NO2i = 300+(NO2-280)*(100/120)
    if(NO2>400):
        NO2i = 400 + (NO2-400)*(100/120)
    return NO2i
```

```python
# Function to calculate O3 (Ozone) individual pollutent index (O3i):
def calculate_O3i(O3):
    O3i = 0
    if (O3<=50):
        O3i = O3*(50/50)
    if(O3>50 and O3<=100):
        O3i = 50+(O3-50)*(50/50)
    if(O3>100 and O3<=168):
        O3i = 100+(O3-100)*(100/68)
    if(O3>168 and O3<208):
        O3i = 200+(O3-168)*(100/(40))
    if(O3>208 and O3<=748):
        O3i = 300+(O3-208)*(100/540)
    if(O3>748):
        O3i = 400 + (O3-748)*(100/540)
    return O3i
```

```python
# Function to calculate CO (Ozone) individual pollutent index (COi):
def calculate_COi(CO):
    COi = 0
    if (CO<=1):
        COi = CO*(50/1)
    if(CO>1 and CO<=2):
        COi = 50+(CO-1)*(50/1)
    if(CO>2 and CO<=10):
        COi = 100+(CO-2)*(100/8)
    if(CO>10 and CO<17):
        COi = 200+(CO-10)*(100/(7))
    if(CO>17 and CO<=34):
        COi = 300+(CO-17)*(100/17)
    if(CO>34):
        COi = 400 + (CO-34)*(100/17)
    return COi
```

```python
# Function to calculate NH3 individual pollutent index (NH3i):
def calculate_NH3i(NH3):
    NH3i = 0
    if (NH3<=200):
        NH3i = NH3*(50/200)
    if(NH3>200 and NH3<=400):
        NH3i = 50+(NH3-200)*(50/200)
    if(NH3>400 and NH3<=800):
        NH3i = 100+(NH3-400)*(100/400)
    if(NH3>800 and NH3<=1200):
        NH3i = 200+(NH3-800)*(100/(400))
    if(NH3>1200 and NH3<=1800):
        NH3i = 300+(NH3-1200)*(100/600)
    if(NH3>1800):
        NH3i = 400 + (NH3-1800)*(100/600)
    return NH3i
```

```python
data_req['SO2'].head()
```

```python
data_req['SO2i'] = data_req['SO2'].apply(calculate_SO2i)
data_req['PM10i'] = data_req['PM10'].apply(calculate_PM10i)
data_req['PM2.5i'] = data_req['PM2.5'].apply(calculate_PM2_5i)
data_req['NH3i'] = data_req['NH3'].apply(calculate_NH3i)
data_req['COi'] = data_req['CO'].apply(calculate_COi)
data_req['Ozonei'] = data_req['Ozone'].apply(calculate_O3i)
data_req['NO2i'] = data_req['NO2'].apply(calculate_NO2i)
```

```python
data_req.head()
```

```python
data_req['AQI'] = data_req[['SO2i', 'PM10i', 'PM2.5i', 'NH3i', 'COi', 'Ozonei', 'NO2i']].max(axis=1)
data_req.head()
```
